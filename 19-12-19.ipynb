```python
# print the digits and count of a number
void main()
{
    int n,r,s=0,count=0;
    scanf("%d",&n);
    while(n!=0)
    {
       count++;
    r=n%10;
    s=s+r;
    n=n/10;
    }
      printf("%d %d",s,count);
}
```

```python
# palindrome or not
void main()
{
    int t,n,r,s=0;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
       
    r=t%10;
    s=s*10+r;
    t=t/10;
    }
      if(n==s)
      printf("palidrome");
      else
      printf("not");
}
```

```python
#armstrong or not
void main()
{
    int t,n,r,s=0;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
       
    r=t%10;
    s=s+r*r*r;
    t=t/10;
    }
      if(n==s)
      printf("armstrong");
      else
      printf("not");
}
```

```python
# armstrong 
#include<math.h>
void main()
{
    int t,n,r,s=0,c;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
       c++;
       t=t/10;
    }
    t=n;
    while(t!=0)
    {
    r=t%10;
    s=s+pow(r,c);
    t=t/10;
    }
      if(n==s)
      printf("armstrong");
      else
      printf("not");
}
```

```python
# armstrong range
#include<math.h>
void main()
{
    int t,n,r,s,c,range;
    scanf("%d",&range);
    for(n=1;n<=range;n++)
    {
    t=n;
    s=0;
    c=0;
    while(t!=0)
    {
       c++;
       t=t/10;
    }
    t=n;
    while(t!=0)
    {
    r=t%10;
    s=s+pow(r,c);
    t=t/10;
    }
      if(n==s)
      printf("  %d  ",n);
}
}
```

```python
#nth palindrome
void main()
{
    int t,n,r,s,c=0,k;
    scanf("%d",&k);
    for(n=0;c<k;n++)
    {
    t=n;
    s=0;
    while(t!=0)
    {
    r=t%10;
    s=s*10+r;
    t=t/10;
    }
      if(n==s)
      {
          c++;
          printf(" %d = %d ",c,n);
      }
    }
}
```
